```python
df=df.replace("Yes",1)
```

```python
df= df.replace("Negative",0)
```

```python
#we"ll consider Gender column as isMale
```

```python
df= df.replace("Male",1)
```

```python
# The columns are not consistently cased
```

```python
# Replace Gender to isMale
```

```python
replace= {"Gender":"isMale"}
```

```python
df=df.rename(columns=replace)
```

```python
df
```

```python
#lowercase everything in the columns
```

```python
df.columns
```

```python
df.columns=df.columns.str.lower()
```
